```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
```

```python
df = pd.read_csv("/content/sudan.csv")
df["Country"] = "Sudan"
```

```python
df["Date"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
df["Month"] = df["Date"].dt.month
```

```python
df.replace(-999, np.nan, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
cols = ["T2M","T2M_MAX","T2M_MIN","PRECTOTCORR","RH2M","WS2M","WS2M_MAX"]
z = df[cols].apply(zscore)
```

```python
df = df.dropna(thresh=int(len(df.columns)*0.7))
df.fillna(method="ffill", inplace=True)
```

```python
df.groupby("Month")["T2M"].mean().plot()
plt.title("Sudan Temperature Trend")
plt.show()
```

```python
df.groupby("Month")["PRECTOTCORR"].sum().plot(kind="bar")
plt.title("Sudan Rainfall Trend")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True))
plt.title("Sudan Correlation")
plt.show()
```
