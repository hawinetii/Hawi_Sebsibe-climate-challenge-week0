```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/kenya.csv")
```

```python
df["Date"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
df["Month"] = df["Date"].dt.month
```

```python
df.replace(-999, pd.NA, inplace=True)
df.fillna(method="ffill", inplace=True)
```

```python
df.groupby("Month")["T2M"].mean().plot()
plt.title("Kenya Temperature Trend")
plt.show()
```

```python
df.groupby("Month")["PRECTOTCORR"].sum().plot(kind="bar")
plt.title("Kenya Rainfall Pattern")
plt.show()
```

```python
sns.heatmap(df.corr())
plt.title("Kenya Correlation")
plt.show()
```
